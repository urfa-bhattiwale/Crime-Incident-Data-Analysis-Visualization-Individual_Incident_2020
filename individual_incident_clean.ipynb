```python
df.dropna(inplace=True)          # Drop missing rows
```

```python
df.fillna(value=0, inplace=True) # Fill missing with 0
```

```python
df.drop_duplicates(inplace=True) # Drop duplicates
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True).fillna(0)
```

```python
df.to_csv(r"D:\Individual_Incident_2020.csv", index=False)
```

```python
# High-Level Incident & Violence Summary
# Total Incidents, Violent Incidents, and Violence Rate
total_incidents = len(df)
violent_incidents = df[df['violence_offense'] > 0].shape[0]
violence_rate = (violent_incidents / total_incidents) * 100

print(f"Total Incidents Recorded: {total_incidents}")
print(f"Incidents Involving Violence: {violent_incidents}")
print(f"Overall Violence Rate: {violence_rate:.2f}%")
```

```python
# Risk Factor Analysis (Guns & Drugs)
# Gun and Drug Involvement Rates
incidents_with_guns = df[df['gun_involvement'] > 0].shape[0]
incidents_with_drugs = df[df['drug_involvement'] > 0].shape[0]

gun_rate = (incidents_with_guns / total_incidents) * 100
drug_rate = (incidents_with_drugs / total_incidents) * 100

print(f"Incidents with Gun Involvement: {incidents_with_guns} ({gun_rate:.2f}%)")
print(f"Incidents with Drug Involvement: {incidents_with_drugs} ({drug_rate:.2f}%)")
```

```python
# Victim Demographics (Minor vs. Non-Minor)
# Total Victims and Minor Victim Proportion
total_victims = df['total_victim'].sum()
minor_victims = df['minor_victim'].sum()
minor_victim_pct = (minor_victims / total_victims) * 100

print(f"Total Number of Victims: {total_victims}")
print(f"Total Minor Victims: {minor_victims}")
print(f"Percentage of Victims who are Minors: {minor_victim_pct:.2f}%")
```

```python
# Offender Relationship Insight
# Familiar vs. Unknown Offender Breakdown
family_offenses = df[df['offender_wi_family'] > 0].shape[0]
unknown_offenders = df[df['offender_not_known'] > 0].shape[0]

family_pct = (family_offenses / total_incidents) * 100
unknown_pct = (unknown_offenders / total_incidents) * 100

print(f"Incidents within the Family: {family_offenses} ({family_pct:.2f}%)")
print(f"Incidents where Offender is Unknown: {unknown_offenders} ({unknown_pct:.2f}%)")
```

```python
#Hourly Crime Pattern
#Insight: Peak crime hours

hourly_crime = df.groupby('hour')['total_offense'].sum()

plt.figure(figsize=(12, 6))
hourly_crime.plot(kind='line')
plt.xlabel("Hour of Day")
plt.ylabel("Total Offenses")
plt.title("Hourly Crime Pattern")
plt.xticks(rotation=0) # Set rotation to 0 for horizontal labels
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

peak_crime_hours = hourly_crime[hourly_crime == hourly_crime.max()]
print(f"Peak Crime Hour(s) and Total Offenses:\n{peak_crime_hours}")
```

```python
import matplotlib.pyplot as plt

victim_demo = df[['male_victim', 'female_victim',
                  'minor_victim', 'non_minor_victim']].sum()

plt.figure()

colors = ['skyblue', 'lightcoral', 'gold', 'lightgreen']

plt.pie(victim_demo,
        labels=victim_demo.index,
        autopct='%1.1f%%',
        colors=colors)

plt.title("Victim Demographics Distribution")
plt.show()
```

```python
#


# Select numeric columns
corr_columns = ['total_offense', 'total_victim', 
                'total_offender', 'property_value']

# Create correlation matrix
corr_matrix = df[corr_columns].corr()

print("Correlation Matrix:")
print(corr_matrix)

# Get correlation values with total_offense
corr_with_offense = corr_matrix['total_offense'].drop('total_offense')

# Plot clustered column chart
plt.figure()
corr_with_offense.plot(kind='bar')

plt.title("Correlation with Total Offense")
plt.ylabel("Correlation Value")
plt.xlabel("Variables")
plt.xticks(rotation=45)

plt.show()

```
